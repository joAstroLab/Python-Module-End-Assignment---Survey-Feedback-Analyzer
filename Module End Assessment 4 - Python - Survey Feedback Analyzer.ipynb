Step 3: Text cleaning:

```python
text_cleaning = []
for tc in feedback_data['Feedback']:
# 1. Remove punctuation (., ,, !, ?):
    for p in [".", ",", "!", "?"]:
      tc = tc.replace(p, "")

# 2. Replace multiple spaces with a single space:
    tc = ' '.join(tc.split())

#3. Remove leading and trailing spaces:
    tc = tc.strip()

#4. Convert all text to lowercase:
    tc = tc.lower()

    text_cleaning.append(tc)

print("Feedback text cleaned successfully!")

print(feedback_data)
```

step 4: Word Count Insights:

```python
# Create a function:
def count_word_in_feedbacks(word):
    word = word.lower()
    count = 0

    for tc in feedback_data['Feedback']:
# Converting feedback to lowercase for case insensitive match:
        tc_lower = tc.lower()

# Check if word exists in feedback (split into words for exact match):
        if word in tc_lower.split():
            count += 1

    return count

# Use this function to print:
print("Word count insights: ")
print("Count of 'good':", count_word_in_feedbacks("good"))
print("Count of 'poor':", count_word_in_feedbacks("poor"))
print("Count of 'excellent':", count_word_in_feedbacks("excellent"))

```

```python
# Print the average rating from all feedbacks:
ratings = feedback_data['Rating']
average_rating = sum(ratings) / len(ratings)

print("Average Rating from all feedbacks:", round(average_rating, 2))
```

```python
# Find and display the feedback with the longest comment:
longest_feedback = ""
max_words = 0

for feedback in feedback_data['Feedback']:
    word_count = len(feedback.split())
    if word_count > max_words:
        max_words = word_count
        longest_feedback = feedback

print("\nLongest Feedback:")
print(longest_feedback)
print("Word Count:", max_words)
```

```python
# Print the list of unique words used across all feedbacks:
unique_words = set()

for feedback in feedback_data['Feedback']:
    words = feedback.split()
    for word in words:
        unique_words.add(word)

print("\nUnique Words Used:")
print(unique_words)
```
